```python
f = lambda x: np.sin(x[0])**2 + np.sin(x[1]) ** 2

def grad_descent(lr, num_iter=100):
    cur_x = np.array([-1., 1.5])
    steps = []
    
    for iter_num in range(num_iter):
        steps.append([cur_x[0], cur_x[1], f(cur_x)])
        grad = np.zeros_like(cur_x)
        
        grad[0] = 2 * np.sin(cur_x[0]) * np.cos(cur_x[0])
        grad[1] = 2 * np.sin(cur_x[1]) * np.cos(cur_x[1])
                    
        cur_x -= lr * grad
    return np.array(steps)
```

```python
steps = grad_descent(0.5, 10)
steps
```
